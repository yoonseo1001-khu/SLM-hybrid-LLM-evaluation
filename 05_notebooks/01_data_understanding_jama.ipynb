## Data Structure

### Columns 설명
- question: 환자 케이스 기반 문제
- choices: 선택지
- answer: 정답
- explanation: 정답 근거 (있을 경우)

### 특징
- question 길이가 매우 김
- clinical 정보 포함 (나이, 증상, 검사 결과)

```python
#길이 분석
df_jama["question_len"] = df_jama["question"].apply(lambda x: len(str(x).split()))

df_jama["question_len"].describe()
```

```python
#선택지 개수 확인
def count_choices(x):
    if isinstance(x, list):
        return len(x)
    return None

df_jama["num_choices"] = df_jama["choices"].apply(count_choices)

df_jama["num_choices"].value_counts()
```
