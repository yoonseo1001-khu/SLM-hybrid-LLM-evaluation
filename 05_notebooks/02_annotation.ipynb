```python
import pandas as pd
from datasets import load_from_disk

# MedQA
medqa_path = "/gpfs/data/oermannlab/gaifl/data/MedQA/med_qa_en_4options_bigbio_qa"
medqa = load_from_disk(medqa_path)
df_medqa = pd.DataFrame(medqa["train"])

# JAMA
jama_path = "/gpfs/data/oermannlab/gaifl/data/jama_clinical_cases.json"
df_jama = pd.read_json(jama_path)
```

```python
# MedQA difficulty labeling
def label_medqa(row):
    q_len = len(str(row["question"]).split())

    if q_len < 12:
        return "easy"
    elif q_len < 25:
        return "medium"
    else:
        return "hard"

df_medqa["difficulty"] = df_medqa.apply(label_medqa, axis=1)
```

```python
#Jama difficulty labeling
def label_jama(row):
    q_len = len(str(row["question"]).split())

    if q_len < 40:
        return "easy"
    elif q_len < 80:
        return "medium"
    else:
        return "hard"

df_jama["difficulty"] = df_jama.apply(label_jama, axis=1)
```

```python
#결과 저장
df_medqa.to_csv("../00_data/annotated/medqa/medqa_annotated.csv", index=False)
df_jama.to_csv("../00_data/annotated/jama/jama_annotated.csv", index=False)
```
