```python
#Load
from sklearn.model_selection import train_test_split

df_medqa = pd.read_csv("../00_data/annotated/medqa/medqa_annotated.csv")
```

## Dataset Split

We split the dataset into train, validation, test, and calibration sets.

- Train: used for model training
- Validation: used for prompt tuning
- Test: used for final evaluation
- Calibration: used for uncertainty estimation

We use stratified sampling to preserve the difficulty distribution.

```python
#Split
# small class 제거
counts = df_medqa["difficulty"].value_counts()
df_medqa = df_medqa[df_medqa["difficulty"].isin(counts[counts >= 2].index)]

# split
train, temp = train_test_split(
    df_medqa,
    test_size=0.3,
    stratify=df_medqa["difficulty"],
    random_state=42
)

val, temp = train_test_split(
    temp,
    test_size=2/3,
    random_state=42
)

test, calibration = train_test_split(
    temp,
    test_size=0.5,
    random_state=42
)
```

```python
#결과 저장
train.to_csv("../00_data/splits/medqa/train.csv", index=False)
val.to_csv("../00_data/splits/medqa/val.csv", index=False)
test.to_csv("../00_data/splits/medqa/test.csv", index=False)
calibration.to_csv("../00_data/splits/medqa/calibration.csv", index=False)
```
