```python
import pandas as pd

df_medqa = pd.read_csv("../00_data/annotated/medqa/medqa_annotated.csv")
```

```python
df_medqa = df_medqa.sample(frac=1, random_state=42).reset_index(drop=True)

n = len(df_medqa)

df_medqa["difficulty"] = "easy"
df_medqa.loc[int(n*0.3):int(n*0.7), "difficulty"] = "medium"
df_medqa.loc[int(n*0.7):, "difficulty"] = "hard"
```

```python
print(df_medqa["difficulty"].value_counts())
```

```python
df_medqa.to_csv("../00_data/annotated/medqa/medqa_annotated_fixed.csv", index=False)
```
