```python
#Load
from sklearn.model_selection import train_test_split

df_medqa = pd.read_csv("../00_data/annotated/medqa/medqa_annotated_fixed.csv")
```

```python
from sklearn.model_selection import train_test_split

# small class 제거 (지금은 사실 필요 없음)
counts = df_medqa["difficulty"].value_counts()
df_medqa = df_medqa[df_medqa["difficulty"].isin(counts[counts >= 2].index)]

# 1차 split (train 70%, temp 30%)
train, temp = train_test_split(
    df_medqa,
    test_size=0.3,
    stratify=df_medqa["difficulty"],
    random_state=42
)

# 2차 split (val / test / calibration)
val, temp2 = train_test_split(
    temp,
    test_size=2/3,
    stratify=temp["difficulty"],   # ⭐ 추가
    random_state=42
)

test, calibration = train_test_split(
    temp2,
    test_size=0.5,
    stratify=temp2["difficulty"],  # ⭐ 추가
    random_state=42
)
```
